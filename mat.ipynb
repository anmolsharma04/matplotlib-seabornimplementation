```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defining data
months = np.array(['jan','feb','march','april','may','june','july','aug','sept','oct','nov','dec'])
sales = np.array([200,300,220,560,890,345,123,563,76,987,123,220])
```

```python
# 1. basic line plot

plt.figure(figsize=(6,4))
plt.plot(months,sales,marker='o',linestyle='-',color='blue') #lineplot of sales over months
plt.title('monthly sales data')
plt.xlabel('months') #label for x-axis
plt.ylabel('sales') #label for y-axis
plt.grid(True) #shows grid lines
plt.show() #displays the plot
```

```python
# 2. histogram 
np.random.seed(0) #for reproducibility
data = np.random.normal(loc=0 , scale=1 , size=1000) #generate random data with normal distribution

plt.figure(figsize=(6,4))
plt.hist(data , bins=30 , color='purple' , edgecolor = 'black') #histogram showing distribution of data
plt.title('RANDOM DATA')
plt.xlabel('value')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```

```python
#3.SCATTER PLOT
np.random.seed(1)
x = np.random.rand(50) * 100 #random x values
y = np.random.rand(50) * 100 #random y values

plt.figure(figsize=(6,4))
plt.scatter(x , y , color='red',alpha =0.7)
plt.title('RANDOM SCATTER PLOT')
plt.xlabel(' X value')
plt.ylabel('Y value')
plt.grid(True)
plt.show()
```

```python
#4. BAR PLOT 
categories = np.array(['A','B','C','D','E'])
value = np.array([10,20,15,25,30])


plt.figure(figsize=(6,4))
plt.bar(categories , value , color=['blue','green','red','purple','orange'])
plt.title('CATEGORY VALUES')
plt.xlabel('category')
plt.ylabel('value')
plt.grid(True)
plt.show()
```

```python
#5 . BOX PLOT
np.random.seed(2)
group1 = np.random.normal(loc=0 , scale=1 , size = 100) #data for group 1
group2 = np.random.normal(loc=1 , scale=1.5 , size = 100) #data for group 2

plt.figure(figsize=(6,4))
plt.boxplot([group1 , group2] , labels=['group1' , 'group2'])
plt.title('BOX PLOT OF TWO GROUPS')
plt.xlabel('group')
plt.ylabel('value')
plt.grid(True)
plt.show()

                          
```

```python
# Heatmap (Correlation Matrix)
# Create a correlation matrix for demonstration
np.random.seed(3)  # For reproducibility
data_matrix = np.random.rand(10, 10)  # Random data matrix

plt.figure(figsize=(6, 4))
plt.imshow(data_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Value')
plt.title('Heatmap of Random Data Matrix')
plt.show()
```

```python
# Generate some simple data using arrays
x = np.linspace(0, 10, 100)  # 100 values from 0 to 10
y1 = np.sin(x)  # Sine wave
y2 = np.cos(x)  # Cosine wave

# Create a figure
plt.figure(figsize=(6, 2))

# First subplot: Sine wave (assigned to the first plot in the grid)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.plot(x, y1, 'b-')  # Line plot for sine wave
plt.title('Sine Wave')  # Title of the subplot

# Second subplot: Cosine wave (assigned to the second plot in the grid)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
plt.plot(x, y2, 'r--')  # Line plot for cosine wave
plt.title('Cosine Wave')  # Title of the subplot

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
# CREATING PLOTS USING SEABORN
import seaborn as sns
import pandas as pd
import warnings #prevents printing of warnings in output
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')

#load the data set
df = pd.read_csv(r'C:\\Users\\anmol\\Downloads\\sales.csv', encoding='ISO-8859-1')
df.head()


#set the style for plots for better visualization
sns.set(style='whitegrid') 
df.head()
```

```python
#1 . HISTOGRAM WITH KDE
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins = 20 , kde = True , color = 'purple')#histogram with kde overlay 
plt.title('Distribution')
plt.xlabel('Age')
plt.ylabel('User_ID')
plt.show()
```

```python
#  2 . SCATTER PLOT WITH HUE 
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age' , y='User_ID' , data = df , hue = 'Gender' , palette = 'deep' , alpha = 0.7) #hue is the differentiating factor 
plt.title('AGE VS USER_ID')
plt.xlabel('Age')
plt.ylabel('User_ID')
plt.show()

#palette='deep' specifies a predefined color palette from Seaborn that assigns different colors to each unique value in the hue variable (in this case, Gender).
#alpha=0.7 means the points will have 70% opacity, making them slightly transparent. This is useful when there are overlapping points, as it allows you to see through overlapping layers, improving visual clarity.
```

```python
# 3 . Box PLOT 
plt.figure(figsize=(6,4))
sns.boxplot(x='Age' , y='User_ID' , data = df , palette = 'pastel' ) #hue is the differentiating factor 
plt.title('AGE VS USER_ID')
plt.xlabel('Age')
plt.ylabel('User_ID')
plt.show()
```

```python
# 4 . BAR PLOT 
plt.figure(figsize=(6,4))
sns.barplot(x='Age' , y='User_ID' , data = df , estimator = 'mean' ,  palette = 'pastel' ) #hue is the differentiating factor 
plt.title('AGE VS USER_ID')
plt.xlabel('Age')
plt.ylabel('User_ID')
plt.xticks(rotation = 45) #rotate x-axis labels for better readability
plt.show()
```

```python
# 5 . PAIR PLOT 
plt.figure(figsize=(6,4))
sns.pairplot(df[['Age' , 'User_ID' , 'Gender']] , diag_kind = 'kde' , hue = 'Gender')
plt.show()
```

```python
# Create the figure
plt.figure(figsize=(8, 6))

# Draw the heatmap using seaborn
# Create a pivot table to count the number of User_ID per Age and Gender
pivot_table = df.pivot_table(index='Age', columns='Gender', values='User_ID', aggfunc='count')

sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt='g')  # annot=True adds values to the cells

# Add labels and title
plt.title('Heatmap: Age vs. User_ID Count')
plt.xlabel('User_ID Count')
plt.ylabel('Age')

# Show the plot
plt.show()
```
